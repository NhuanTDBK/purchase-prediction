## New users every month

```python
init_group = set()
new_users = []
for name,group in df_flatten.resample('M', on='date')["csn"]:
    new_members = 0
    if len(init_group) == 0:
        for t in group.values:
            init_group.add(t)
    else:
        for t in group.values:
            if t not in init_group:
                new_members+=1
                init_group.add(t)
    new_users.append([name,new_members])
new_users = np.array(new_users)
df_new_users = pd.DataFrame(new_users,columns=["date","New users"])
```

```python
df_new_users["date"] = pd.to_datetime(df_new_users["date"])
```

```python
%matplotlib inline
ax = plt.subplot()
ax.set_ylabel("Distinct users")
df_new_users.set_index(["date"]).join(df_flatten.resample('M', on='date')["csn"].nunique())\
    .plot(ax=ax,kind="bar",title="New users per month",figsize=(30,10))
```

## Reorder pattern
We assume hypothesis is that people have regular pattern to buy things => good indicator to predict next purchase

```python
last_purchases_user_items = {}
days_prior_purchase = {}
```

```python
for user_id in dat["csn"].unique():
    last_purchases_user_items[user_id] = {}
    days_prior_purchase[user_id] = {}
```

```python
def calculate_last_purchase(dataset):
    last_purchases_user = {}
    priors_user_day = np.zeros(len(dataset),dtype=np.uint)
    idx = 0
    for _, row in dataset.iterrows():
        user_id = row["csn"]
        date = row["date"]
        delta = 0 
        if user_id in last_purchases_user:
            timestamp = row["date"] 
            delta = (timestamp - last_purchases_user[user_id]).days
        last_purchases_user[user_id] = date
        priors_user_day[idx] = delta
        idx+=1
    return priors_user_day
```

```python
priors_user_day = calculate_last_purchase(dat)
dat["user_comeback_after"] = priors_user_day
onetime_users_id = dat["csn"].iloc[np.where(dat["csn"].value_counts()==1)[0]]
```

```python
%matplotlib inline
ax = dat[(dat["user_comeback_after"]>0)&(~dat["csn"].isin(onetime_users))]\
    .groupby(["user_comeback_after"])["csn"].nunique()\
    .head(100)\
    .plot(kind="bar",figsize=(20,10),title="How many days customers comeback")
ax.set_ylabel("Number of customers")
ax.set_xlabel("Days")
```

```python
d = dat.groupby(["csn","date"])['day_of_week'].count().reset_index(name="count_date").query("count_date>1")
if len(d) == 0:
    print("No customers ordered on same days")
```

=> Pattern mua hàng lặp lại sau 1 ngày, 7 ngày, 14 ngày, 21 ngày, 30 ngày

```python
def calculate_prior_items(data):
    last_purchases_user_items = {}
    items_priors_day = np.zeros(len(data),dtype="uint")
    for user_id in data["csn"].unique():
        last_purchases_user_items[user_id] = {}
        
    for idx, row in data.sort_values(by=["date"],ascending=True).iterrows():
        user_id = row["csn"]
        item_id = row["article"]
        timestamp = row["date"]
        delta = 0
        if item_id in last_purchases_user_items[user_id]:        
            last_time_purchase = last_purchases_user_items[user_id][item_id]
            delta = (timestamp - last_time_purchase).days
        last_purchases_user_items[user_id][item_id] = timestamp
        items_priors_day[idx] = delta
    return items_priors_day
```

```python
df_flatten["days_user_items_prior_purchase"] = calculate_prior_items(df_flatten)
```

```python
%matplotlib inline
ax = plt.subplot()
df_flatten.query("days_user_items_prior_purchase>0").groupby(["days_user_items_prior_purchase"])["csn"].nunique()\
    .plot(ax=ax,title="Items reordered pattern",kind="bar",figsize=(30,10),fontsize=14)
```

=> Pattern sp được mua nhắc lại sau 7 ngày, 14 ngày, 21 ngày, 30 ngày

## How many items that are reordered?

### 75% of items are reordered

```python
df_reorder = df_flatten.groupby(["article"])["days_user_items_prior_purchase"].max().reset_index(name="max_reorder")
```

```python
df_reorder["reorder"] = df_reorder["max_reorder"].map(lambda d: 1 if d > 0 else 0)
```

```python
t = df_reorder.groupby(["reorder"])["max_reorder"].count().reset_index(name="val")
t["ratio"] = t["val"]*100/len(df_reorder)
print(t.head())
del t
```
